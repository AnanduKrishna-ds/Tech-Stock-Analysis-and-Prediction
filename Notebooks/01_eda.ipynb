# Introduction

### Exploratory Data Analysis of Selected Technology Stocks

This project focuses on analyzing historical stock market data for five major technology companies: **NVDA (NVIDIA), INTC (Intel), AAPL (Apple), TSLA (Tesla), and NFLX (Netflix)**. These companies represent different segments within the technology sector, including semiconductors, consumer electronics, electric vehicles, and digital streaming, making them suitable for comparative financial analysis.

The dataset was obtained using the `yfinance` library and covers the period from **December 30, 2017 to December 31, 2025**. It contains daily trading information for each stock, including Open, High, Low, Close, Adjusted Close prices, and Volume. Since multiple stocks were downloaded simultaneously, the dataset is structured with a MultiIndex column format, where price attributes are grouped under each ticker symbol.

This exploratory data analysis aims to examine the behavior and performance of these stocks over time. The analysis will explore overall price trends, return patterns, volatility characteristics, trading activity, and inter-stock relationships. By studying these aspects, the project seeks to develop a deeper understanding of market dynamics and establish a solid foundation for potential financial modeling or predictive analysis.

This EDA serves as the initial step in transforming raw financial time-series data into meaningful insights through visualization, statistical summaries, and structured analysis.

# Univariate Analysis

```python
data['year']=data['date'].dt.year
data['month']=data['date'].dt.month_name()
data['day']=data['date'].dt.day_name()
```

### 	How have individual stock prices evolved over time during the selected period?

```python
# melt the wide format to long format
data2 = data.reset_index()
adj_close_cols = [col for col in data2.columns if col.startswith("adj close")]
data_long = data2.melt(id_vars='date',value_vars=adj_close_cols,var_name='Stock',value_name='Adj Close')
data_long['Stock'] = data_long['Stock'].str.replace('adj close ', '', case=False).str.strip()

g = sns.FacetGrid(data_long, col='Stock', col_wrap=1, height=4,aspect=4,sharey=False)
g.map(sns.lineplot, 'date', 'Adj Close')
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Date", "Adjusted Close")
plt.suptitle('Stock Price Evolution Over Time', y=1.02)
plt.show()
```

##### The stock price evolution plots show NVDA delivered the strongest long-term growth, rising sharply from low levels in 2018–2019 to well above $180 by 2025–2026 — clearly the top performer in percentage terms. TSLA exhibited the highest volatility with sharp rallies and deep pullbacks, yet achieved substantial net gains, reaching peaks near $480. AAPL displayed steady and consistent upward movement with moderate corrections along the way. NFLX experienced a significant decline around 2022 but later recovered and resumed its upward trend. INTC was the weakest performer, facing prolonged declines and only a partial recovery toward the end of the period. Overall, NVDA led decisively in overall appreciation, followed by TSLA, while INTC lagged behind the group.


### 	What are the overall return characteristics of each stock?

```python
data_long['Returns']=data_long['Adj Close'].pct_change()
data_long.dropna(inplace=True)
data_long.head()
sns.relplot(data=data_long,x="date",y='Returns',col='Stock',col_wrap=1,height=4,aspect=4,kind="line")
plt.show()
print(data_long['Returns'].describe().T)
```

##### The daily returns plots indicate that all stocks experienced periods of heightened volatility, with sharp negative spikes visible in the early years and around major market corrections. TSLA stands out with the most frequent and extreme positive and negative swings, highlighting its high-risk, high-reward behavior. NVDA also shows increasing return dispersion in later years, aligning with its strong upward momentum. AAPL and NFLX exhibit relatively tighter clustering around zero, suggesting more controlled daily fluctuations. INTC shows occasional sharp drops but remains less explosive compared to TSLA and NVDA. Overall, TSLA demonstrates the highest return volatility, while AAPL appears comparatively more stable in daily movements.

### 	How volatile is each stock, and how does volatility change over time?

```python
data_long['Rolling Std']=data_long.groupby('Stock')['Returns'].transform(lambda x:x.rolling(window=21).std().shift(1))
data_long.dropna(inplace=True)
sns.relplot(data=data_long,x="date",y='Rolling Std',col='Stock',col_wrap=1,height=4,aspect=4,kind="line")
plt.suptitle("Volatility change of Stocks",y=1.02)
plt.show()
```

#####  The rolling 21-day volatility shows TSLA had the highest and most persistent volatility, with frequent sharp spikes across the period. NVDA also experienced notable volatility surges, particularly during strong price rallies. NFLX and INTC displayed occasional spikes but remained moderate overall. AAPL maintained comparatively lower and more stable volatility levels. Overall, TSLA was the most volatile stock, while AAPL showed the most consistent risk profile.

### 	What does the distribution of daily returns reveal about risk and asymmetry?

```python
sns.displot(x="Adj Close",data=data_long,col="Stock",kind="hist",kde=True,col_wrap=2,height=4,aspect=2)
plt.show()
```

##### The price distributions show TSLA and NVDA have the widest right tails, indicating the strongest overall price growth. AAPL demonstrates steady appreciation with moderate spread, while NFLX shows broader dispersion due to both gains and declines. INTC remains concentrated in lower price ranges, reflecting weaker long-term performance. Overall, TSLA and NVDA lead in price expansion, while INTC trails the group.

### 	How strongly are these stocks correlated with each other?

```python
data2=data2[['adj close aapl', 'adj close intc', 'adj close nflx','adj close nvda', 'adj close tsla']]
returns = data2.pct_change().dropna()
corr=returns.corr()
plt.figure(figsize=(9,7))
sns.heatmap(data=corr,cmap="viridis",annot=True)
plt.show()
```

##### The heatmap indicates that all stocks are positively correlated, moving broadly in the same direction. The strongest relationship appears between AAPL and NVDA, while INTC shows the weakest correlations with others. Overall, correlations are moderate rather than high, implying partial diversification benefits within the portfolio.

### 	Do certain stocks consistently exhibit higher risk relative to others?

```python
risk=data_long.groupby('Stock')['Returns'].std()
risk.sort_values().plot(kind='bar')
plt.ylabel("Volatility (Std of Returns)")
plt.ylabel("Volatility (Std of Returns)")
plt.title("Comparitive Risk of Stocks (aapl, intc, nflx, nvda, tsla)")
plt.show()
```

##### The chart shows TSLA has the highest volatility, indicating the greatest risk. NVDA follows with elevated risk, while INTC and NFLX fall in the mid-range. AAPL records the lowest volatility, making it the most stable stock in the sample.

### 	Do relationships between stocks remain stable over time?

```python
returns_wide = data_long.pivot(index='date',columns='Stock',values='Returns').dropna()
rol_col = [["aapl", "tsla"], ["nvda", "nflx"], ["intc", "tsla"]]
fig, axes = plt.subplots(1, 3, figsize=(15, 6))
axe_flat = axes.flatten()
for pair, ax in zip(rol_col, axe_flat):
    stock1, stock2 = pair
    rolling_corr = returns_wide[stock1].rolling(window=120).corr(returns_wide[stock2])
    sns.lineplot(x=rolling_corr.index,y=rolling_corr.values, ax=ax,color='purple',linewidth=2)
    ax.set_title(f"{stock1.upper()} vs {stock2.upper()} (120d Corr)")
    ax.set_ylabel("Rolling Correlation")
    ax.set_xlabel("Date")
    ax.axhline(0, color='gray', linestyle='--', alpha=0.5)
    ax.grid(True, alpha=0.3)
plt.suptitle("Stock Relationship Stability (Rolling Correlations)", fontsize=16)
plt.tight_layout()
plt.show()
```

##### The rolling correlations vary significantly over time, showing that stock relationships are unstable. AAPL–TSLA and NVDA–NFLX maintain mostly moderate positive correlation but with clear shifts across periods. INTC–TSLA exhibits weaker and more volatile correlation levels. Overall, correlations are time-varying, highlighting changing diversification benefits.

###  What was the worst loss an investor would have experienced?

```python
cum_returns = (1 + returns).cumprod()
rolling_max = cum_returns.cummax()
drawdown = (cum_returns - rolling_max) / rolling_max
max_drawdown = drawdown.min()
clean_index = max_drawdown.index.str.replace('adj close ', '').str.upper()
max_drawdown.index = clean_index
max_drawdown.sort_values().plot(kind='bar', figsize=(8,5))
plt.gca().invert_yaxis()
plt.title("Maximum Drawdown Comparison")
plt.xticks(rotation=0)
plt.ylabel("Max Drawdown")
plt.show()
```

##### The chart shows TSLA and INTC recorded the largest maximum drawdowns, indicating the deepest peak-to-trough losses. NVDA experienced a moderate decline, while AAPL and NFLX had relatively smaller drawdowns. Overall, TSLA and INTC carried the highest downside risk during the period.

###  Does diversification actually reduce volatility?

```python
individual_vol = returns.std()
weights = np.ones(len(returns.columns)) / len(returns.columns)
portfolio_return = returns.dot(weights)
portfolio_vol = portfolio_return.std()
comparison = individual_vol.copy()
comparison['Portfolio'] = portfolio_vol
plot_vol = individual_vol.copy()
plot_vol['Portfolio'] = portfolio_vol
plot_vol.index = plot_vol.index.str.replace('adj close ', '').str.upper()
plot_vol.sort_values().plot(kind='bar', figsize=(8,5))
plt.title("Volatility Comparison: Individual vs Portfolio")
plt.ylabel("Standard Deviation of Returns")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

##### The equal-weight portfolio shows lower volatility than most individual stocks, demonstrating effective risk reduction through diversification despite positive correlations among the assets.

# Outlier Analysis

```python
g = sns.FacetGrid(data_long, col='Stock', col_wrap=2, height=4, sharey=False)
g.map(sns.boxplot, 'Stock', 'Returns')
g.set_titles(col_template="{col_name}")
g.set_axis_labels("Stock", "Returns")
plt.suptitle('Return Distribution and Outliers per Stock', y=1.02)
plt.show()
```

##### The boxplots indicate clear differences in return volatility across the stocks:

* TSLA and NVDA exhibit the widest interquartile ranges and the longest whiskers, indicating the highest day-to-day variability. They also show numerous extreme outliers on both  the positive and negative sides, reflecting frequent large price swings.

* NFLX shows moderate volatility, with a noticeable number of outliers but a narrower spread than TSLA and NVDA.

* AAPL has a relatively tighter distribution compared to the more volatile growth stocks, though it still displays several significant outliers.

* INTC appears to have the smallest interquartile range and shorter whiskers, suggesting comparatively lower volatility among the group.

* Overall, while all stocks experience occasional extreme returns, TSLA and NVDA stand out as the most volatile, and INTC as the most stable in this sample.

```python
def find_outliers(group):
    Q1 = group['Returns'].quantile(0.25)
    Q3 = group['Returns'].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return group[(group['Returns'] < lower) | (group['Returns'] > upper)]
outliers = data_long.groupby('Stock').apply(find_outliers).reset_index(drop=True)
```

```python
print(f"Total outliers found: {len(outliers)}\n")

for stock in outliers['Stock'].unique():
    stock_outliers = outliers[outliers['Stock'] == stock][['date', 'Returns']].sort_values('date')
    print(f"Stock: {stock.upper()} — {len(stock_outliers)} outliers")
    print(stock_outliers.to_string(index=False))
    print()
```

```python
fig, axes = plt.subplots(len(outliers['Stock'].unique()), 1, 
                          figsize=(12, 4 * len(outliers['Stock'].unique())))

for ax, stock in zip(axes, outliers['Stock'].unique()):
    stock_outliers = outliers[outliers['Stock'] == stock].sort_values('date')
    
    colors = ['red' if x < 0 else 'green' for x in stock_outliers['Returns']]
    
    ax.bar(range(len(stock_outliers)), stock_outliers['Returns'], 
           color=colors, width=0.6)
    
    ax.set_xticks(range(len(stock_outliers)))
    ax.set_xticklabels(stock_outliers['date'].dt.strftime('%Y-%m-%d'), 
                       rotation=90, fontsize=7)
    
    ax.axhline(y=0, color='black', linestyle='--', linewidth=0.8)
    ax.set_title(f'{stock.upper()} — Outlier Returns')
    ax.set_xlabel('Date')
    ax.set_ylabel('Returns')

plt.suptitle('Significant Market Events per Stock', y=1.01)
plt.tight_layout()
plt.show()
```

#### Outlier Analysis — Significant Market Events

**Key observations:**

* INTC shows the largest single-day drop (around -26%), indicating a major negative shock
* NVDA has mostly positive outliers, reflecting strong upside momentum periods
* TSLA displays the most frequent large swings in both directions
* NFLX has a few sharp downside events
* Several overlapping dates suggest broader market-wide shocks

**Conclusion:**
These outliers reflect real market events and should be retained for analysis.
